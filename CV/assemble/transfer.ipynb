```python
import os
import shutil
from pathlib import Path

# 假设 CSV 文件和 imgs 文件夹的路径
csv_file_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/train_data.csv'
train_imgs_dir_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/imgs'
train_output_base_dir = 'imgs_processed/train'

# 读取 CSV 文件
with open(csv_file_path, mode='r') as csv_file:
    lines = csv_file.readlines()[1:]  # 跳过标题行

# 确保输出目录存在
Path(train_output_base_dir).mkdir(parents=True, exist_ok=True)

# 遍历 CSV 文件中的每一行
for line in lines:
    # 分割文件名和标签
    filename, label = line.strip().split(',')
    # 创建类别文件夹路径
    label_dir_path = Path(train_output_base_dir) / label
    # 确保类别文件夹存在
    label_dir_path.mkdir(parents=True, exist_ok=True)
    # 构建原始图片路径和目标路径
    src_img_path = Path(train_imgs_dir_path) / filename
    dest_img_path = label_dir_path / filename
    # 移动图片到相应的类别文件夹
    shutil.copy(str(src_img_path), str(dest_img_path))
    print(f'Moved {filename} to {label_dir_path}')

print('Conversion completed.')


# 假设 CSV 文件和 imgs 文件夹的路径
csv_file_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/test_data.csv'
train_imgs_dir_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/imgs'
train_output_base_dir = 'imgs_processed/test'

# 读取 CSV 文件
with open(csv_file_path, mode='r') as csv_file:
    lines = csv_file.readlines()[1:]  # 跳过标题行

# 确保输出目录存在
Path(train_output_base_dir).mkdir(parents=True, exist_ok=True)

# 遍历 CSV 文件中的每一行
for line in lines:
    # 分割文件名和标签
    filename, label = line.strip().split(',')
    # 创建类别文件夹路径
    label_dir_path = Path(train_output_base_dir) / label
    # 确保类别文件夹存在
    label_dir_path.mkdir(parents=True, exist_ok=True)
    # 构建原始图片路径和目标路径
    src_img_path = Path(train_imgs_dir_path) / filename
    dest_img_path = label_dir_path / filename
    # 移动图片到相应的类别文件夹
    shutil.copy(str(src_img_path), str(dest_img_path))
    print(f'Moved {filename} to {label_dir_path}')
    
    
# 假设 CSV 文件和 imgs 文件夹的路径
csv_file_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/val_data.csv'
train_imgs_dir_path = '/home/sml/Deep-Learning-Basic-Code/CV/Fine grained classification of birds/imgs'
train_output_base_dir = 'imgs_processed/val'

# 读取 CSV 文件
with open(csv_file_path, mode='r') as csv_file:
    lines = csv_file.readlines()[1:]  # 跳过标题行

# 确保输出目录存在
Path(train_output_base_dir).mkdir(parents=True, exist_ok=True)

# 遍历 CSV 文件中的每一行
for line in lines:
    # 分割文件名和标签
    filename, label = line.strip().split(',')
    # 创建类别文件夹路径
    label_dir_path = Path(train_output_base_dir) / label
    # 确保类别文件夹存在
    label_dir_path.mkdir(parents=True, exist_ok=True)
    # 构建原始图片路径和目标路径
    src_img_path = Path(train_imgs_dir_path) / filename
    dest_img_path = label_dir_path / filename
    # 移动图片到相应的类别文件夹
    shutil.copy(str(src_img_path), str(dest_img_path))
    print(f'Moved {filename} to {label_dir_path}')
```
